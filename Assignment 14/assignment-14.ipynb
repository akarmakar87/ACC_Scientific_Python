## Texas Only: Cases per County vs Mask Use

When I was looking through the New York Times COVID data, I found two .csv files: one that gave COVID cases by county for all the counties in the US, and one that gave data on "mask use" for all the counties in the US. Basically, people from each county were surveyed on how often they wore masks (NEVER, RARELY, SOMETIMES, FREQUENTLY, ALWAYS). I was curious if there was a visible correlation between the mask use of counties and the number of cases within the counties.

In order to connect the number of cases in a county with the mask use, I had to merge the two datasets on a common attribute: fips, which is a number assigned to each county in the US.

```python
# Create DataFrame with county and case info
counties_cases = pd.read_csv('county-cases.csv')
counties_cases
```

```python
# Create DataFrame with counties and mask use info
counties_mask_use = pd.read_csv('mask-use-by-county.csv')
counties_mask_use.columns.values[0] = 'fips'
counties_mask_use
```

```python
# Merge the two DataFrames
temp_df = pd.merge(counties_cases, counties_mask_use)
temp_df
```

```python
# Create new DataFrame with only Texas counties and the columns I need
mask_cases = temp_df.loc[temp_df['state'] == 'Texas',['county','cases', 'NEVER', 'RARELY', 'SOMETIMES', 'FREQUENTLY', 'ALWAYS']].copy()
mask_cases
```

```python
mask_cases.describe()
```

I decided that I wanted to work with only Texas counties, so I created a new DataFrame with only the counties marked as within Texas.

```python
a = np.array(mask_cases['NEVER'])
b = np.array(mask_cases['cases'])

ax = plt.axes(label='never')
ax.set(xlabel='% of County that NEVER Wears Masks', 
       ylabel='# of COVID Cases Within County', 
       title='Texas Counties: Mask Use vs COVID Cases')

plt.xlim(0, mask_cases['NEVER'].max())
plt.ylim(0, mask_cases['cases'].mean())

plt.scatter(a, b, alpha=0.75, c ='red',s=100, edgecolors='none')
plt.grid(linestyle=":", color='grey')
```

```python
c = np.array(mask_cases['RARELY'])
d = np.array(mask_cases['cases'])

ax = plt.axes(label='rarely')
ax.set(xlabel='% of County that RARELY Wears Masks', 
       ylabel='# of COVID Cases Within County', 
       title='Texas Counties: Mask Use vs COVID Cases')

plt.xlim(0, mask_cases['RARELY'].max())
plt.ylim(0, mask_cases['cases'].mean())

plt.scatter(c, d, alpha=0.75, c ='orange',s=100, edgecolors='none')
plt.grid(linestyle=":", color='grey')
```

```python
e = np.array(mask_cases['SOMETIMES'])
f = np.array(mask_cases['cases'])

ax = plt.axes(label='sometimes')
ax.set(xlabel='% of County that SOMETIMES Wears Masks', 
       ylabel='# of COVID Cases Within County', 
       title='Texas Counties: Mask Use vs COVID Cases')

plt.xlim(0, mask_cases['SOMETIMES'].max())
plt.ylim(0, mask_cases['cases'].mean())

plt.scatter(e, f, alpha=0.75, c ='green',s=100, edgecolors='none')
plt.grid(linestyle=":", color='grey')
```

```python
g = np.array(mask_cases['FREQUENTLY'])
h = np.array(mask_cases['cases'])

ax = plt.axes(label='frequently')
ax.set(xlabel='% of County that FREQUENTLY Wears Masks', 
       ylabel='# of COVID Cases Within County', 
       title='Texas Counties: Mask Use vs COVID Cases')

plt.xlim(0, mask_cases['FREQUENTLY'].max())
plt.ylim(0, mask_cases['cases'].mean())

plt.scatter(g, h, alpha=0.75, c ='blue',s=100, edgecolors='none')
plt.grid(linestyle=":", color='grey')
```

```python
i = np.array(mask_cases['ALWAYS'])
j = np.array(mask_cases['cases'])

ax = plt.axes(label='always')
ax.set(xlabel='% of County that ALWAYS Wears Masks', 
       ylabel='# of COVID Cases Within County', 
       title='Texas Counties: Mask Use vs COVID Cases')

plt.xlim(0, mask_cases['ALWAYS'].max())
plt.ylim(0, mask_cases['cases'].mean())

plt.scatter(i, j, alpha=0.75, c ='purple',s=100, edgecolors='none')
plt.grid(linestyle=":", color='grey')
```

```python
max_val = np.max([mask_cases['NEVER'].max(),
              mask_cases['RARELY'].max(),
              mask_cases['SOMETIMES'].max(),
              mask_cases['FREQUENTLY'].max(),
              mask_cases['ALWAYS'].max()])

ax = plt.axes(label='all')
ax.set(xlabel='Mask Use in County (%)', 
       ylabel='# of COVID Cases Within County', 
       title='Texas Counties: Mask Use vs COVID Cases')

plt.xlim(0, max_val)
plt.ylim(0, mask_cases['cases'].mean())

plt.scatter(a, b, alpha=0.5, c ='red',s=100, edgecolors='none', label="NEVER")
plt.scatter(c, d, alpha=0.5, c ='orange',s=100, edgecolors='none', label="RARELY")
plt.scatter(e, f, alpha=0.5, c ='green',s=100, edgecolors='none', label="SOMETIMES")
plt.scatter(g, h, alpha=0.5, c ='blue',s=100, edgecolors='none', label="FREQUENTLY")
plt.scatter(i, j, alpha=0.5, c ='purple',s=100, edgecolors='none', label="ALWAYS")

plt.grid(linestyle=":", color='grey')
plt.legend(numpoints=1,loc='upper right')
print ("COMBINE ALL THE PREVIOUS SCATTERPLOTS INTO ONE:")
```
