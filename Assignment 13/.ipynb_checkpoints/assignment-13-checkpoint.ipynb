## Requirement 4

```python
# Use the Matplotlib object-oriented interface to display the sine, cosine,tangent, and arctangent of x
import numpy as np,matplotlib.pyplot as plt
x = np.linspace(0,10,100)
fig, ax = plt.subplots(4, figsize=(12,12))

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
ax[2].plot(x, np.tan(x))
ax[3].plot(x, np.arctan(x))
```

## Requirement 7

```python
# Plot the tanh() function using six different color formats. Modify x slightly in each plot so the lines are distinct.

x = np.linspace(0,10,1000)

plt.plot(x, np.tanh(x - 0), color='red')
plt.plot(x, np.tanh(x - 1), color='orange')
plt.plot(x, np.tanh(x - 2), color='yellow')
plt.plot(x, np.tanh(x - 3), color='green')
plt.plot(x, np.tanh(x - 4), color='blue')
plt.plot(x, np.tanh(x - 5), color='purple')
```

```python
# Use RandomState() and plot() to demonstrate 11 different plot marker types. Also, display the markers in a legend.

rng = np.random.RandomState(1234)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(3), rng.rand(3), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8)
```

## Requirement 17

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',cmap='PuOr', alpha=0.5)
plt.colorbar()
```

## Requirement 19

```python
plt.style.use('classic')
%matplotlib inline

fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='green')
ax.axis('equal')

# specify the lines and labels of the first legend
ax.legend(lines[:2], ['line A', 'line B'],
          loc='upper left', frameon=True)

# Create the second legend and add the artist manually.
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower left', frameon=True)
ax.add_artist(leg)
```
